```python
from statistics import mean

print(data["rangeA"].dtype)
#print(data["rangeA"].astype('int64').dtype)
q = data["rangeA"].replace(r'[/-]',',',regex=True)
q[q.isnull()] = 0

#x = q[q.str.find(',') > 0]
#[int(mean(map(lambda x: int(x),e.split(',')))) for e in x]

x = q[q.str.find(',') > 0]
q[q.str.find(',') > 0] = [mean(map(lambda x: int(x),e.split(','))) for e in x]
display(q[q.str.find(',') > 0])

data["rangeClean"] = q.astype('int')
```

```python
cols_no_nulls = data.columns[data.isnull().sum() < 5000]
```

```python
# Clean the invalid columns
data_clean = data[cols_no_nulls] 
```

```python
null_displ = data_clean[ data['displ'].isnull() == True ]
null_displ = null_displ[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
display(null_displ.fillna(0))

```

```python
data_clean[['displ', 'cylinders']].fillna(0, inplace=True)
```

```python
data_clean[ data_clean['cylinders'].isnull() == True]
```

```python
import numpy as np
low_variance = []

for col in data_clean._get_numeric_data():
    minimum = min(data_clean[col])
    ninety_perc = np.percentile(data_clean[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)

print(low_variance)
```

```python
data_clean = data_clean.drop(low_variance, axis=1)
```

```python
data_clean.shape
```

```python
stats = data_clean.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=data.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = data_clean[(data_clean[col] < lower) | 
                   (data_clean[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```

```python
outliers['Outlier'].value_counts()
```

```python
outliers[outliers['comb08U'] > 100] = 50
```
