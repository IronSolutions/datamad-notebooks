```python
import pandas as pd

df = pd.read_csv('./data/vehicles/vehicles.csv')
df.head()
```

```python
pd.crosstab(df['Make'],df["Year"])[[2008,2009,2010]].sort_values(2008, ascending=False)
```

```python
only_audi = df[(df['Make']=="Audi") | (df['Make']=="BMW")]
crt = pd.crosstab([only_audi["Make"], only_audi["Model"]],[only_audi["Year"],only_audi["Fuel Type"]])
display(crt)
```

```python
numeric = df._get_numeric_data()
numeric["Make"] = df["Make"]
numeric.head()
```

```python
# Agrupar marcas y calcular la media de todas las columnas por marca
numeric.groupby("Make").agg(['mean',"max","min"])
```

```python
numeric[["Make","Cylinders"]].groupby("Make").agg(['mean',"max","min"])
```

```python
def aggCustom(group):
    return group.max() - group.mean()

numeric[["Make","Cylinders"]].groupby("Make").agg(aggCustom)
```

```python
def byNumCilindros(group):
    m = group.mean()
    if m <= 4:
        return "Low"
    elif m > 4 and m < 8:
        return "Middle"
    else:
        return "High"

numeric[["Make","Cylinders"]].groupby("Make").agg(byNumCilindros)
```

```python
# Agg solo se puede usar con groupy delante
numeric.groupby("Make").agg({
    "Cylinders":byNumCilindros,
    "Highway MPG":"min",
    "Combined MPG":"max"
})
```
