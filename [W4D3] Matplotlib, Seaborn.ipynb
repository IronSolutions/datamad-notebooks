```python
total_record_time = 10

temperature_sample_frequency = int(1/0.1)
humidity_sample_frequency = 1

temperature = np.random.random(temperature_sample_frequency*total_record_time)
humidity = np.random.random(humidity_sample_frequency*total_record_time)

display(temperature,humidity)
```

```python
from scipy.interpolate import interp1d


newindex = np.arange(0, 10)
display(newindex, humidity)
interpolator = interp1d(newindex, humidity)

newindex = np.arange(0,9,0.1)

interpolated_humidity = interpolator(newindex)

plt.plot(np.arange(0,10,0.1),temperature,'o--')
plt.plot(newindex,interpolated_humidity ,'o-')
plt.ylabel("Temperature")
plt.xlabel("Tiempo")
ax2 = plt.twinx()
ax2.set_ylabel("Humidity")
plt.legend(["Temperature","Humidity Interpolated"])
plt.show()
```

```python
with plt.xkcd():
    plt.figure()
    for i in range(4):
        x = np.random.random(10)
        plt.subplot(4,1,i+1) ## subplot is a matrix of 1 row and 2 columns
        plt.plot(x,'o-', color=np.random.random(3))
        plt.legend([f"A{i}"])                                    
        plt.title(f"A{i}")


    ax = plt.gca()
    plt.figure()
    plt.plot(np.random.random(10))
    ax.plot(np.random.random(10))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")

rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```
