```python
import requests
from bs4 import BeautifulSoup
res = requests.get("https://www.emsc-csem.org/Earthquake/")
soup = BeautifulSoup(res.text)
```

```python
rows = soup.select('tr.normal')

def processRow(row):
    data = row.select('td')
    #display(data)
    return {
    "datetime":str.strip(data[3].select('a')[0].text),
    "lat": float(data[4].text),
    "lng": float(data[6].text),
    "region": str.strip(data[-2].text)
    } 
    
#processRow(rows[0])
datalist = list(map(processRow, rows))
```

```python
import pandas as pd

pd.DataFrame(datalist)
```
